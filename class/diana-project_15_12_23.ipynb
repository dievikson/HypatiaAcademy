```python
# use this to convert euros to USD


import requests

base_currency = 'EUR'
target_currency = 'USD'
# Make a request to the Open Exchange Rates API
url = f'https://open.er-api.com/v6/latest/{base_currency}'
response = requests.get(url)
# Check if the request was successful (status code 200)
if response.status_code == 200:
    data = response.json()
    exchange_rate = data['rates'][target_currency]
    print(f'1 {base_currency} = {exchange_rate} {target_currency}')
else:
    print(f'Error: Unable to fetch exchange rates. Status code: {response.status_code}')
```
