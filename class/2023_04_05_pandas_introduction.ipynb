```python
# where we say look up rows that are in the state of South Carolina


'''
df.loc() means find by lots of different techniques


df['some column name'] == 'Some value']

then sort it

'''


df.loc[df["State"] == "SC"].sort_values(by="County").iloc[20:25]
```

```python
df.info()
```

```python
df.loc[df["Percent Electric Vehicles"] > 0.2]


```

```python
df['Battery Electric Vehicles (BEVs)'] = pd.to_numeric(df['Battery Electric Vehicles (BEVs)'])
```

```python
df.info()
```

```python
df['Percent Electric Vehicles'].describe()
```

```python
# find missing data.  NaN means no data.  so no data will miss up out ML model


# this only checks one column.  to clear up data we would want to delete all rows
# that had any empty columns



df.loc[df['County'].isnull() == True]
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.iloc[:3]
```

```python
# select a subset of column and rows

# do this does select 3 rows and only 3 columns from those rows


df.loc[[4,5,6],['Date','County','State']]
```

```python
# this makes a new dataframe from the two columns Date and County.  But it is very larg
# so we use .loc() to just print a small part of it, rows 0 through 4

df[['Date', 'County']].iloc[0:4]
```

```python
df.loc[:, ['Date', 'County']].iloc[0:4]
```
