```python
import matplotlib.pyplot as plt

a = np.random.normal(loc=0, scale=3, size=25)

b= np.round(a).astype(int)
c= b.reshape(5,5)

plt.imshow(c, cmap='viridis')

plt.colorbar()

for i in range(c.shape[0]):
    for j in range(c.shape[1]):
        plt.annotate(str(c[i, j]), xy=(j, i), ha='center', va='center', color='white')

plt.annotate(str(c[3, 3]),xy=(3,3), ha='center', va='center', color='white')

plt.show()
```

heatmap charts are used for visualizing patterns, trends, and correlations in data by representing values using colors. They provide an intuitive way to identify areas of high and low density in data, making it easier to detect patterns and anomalies.
