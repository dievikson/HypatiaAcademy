```python
import matplotlib.pyplot as plt

# Create a new figure
# this is the plain constructor to make a figure object.  meaning we don't it anything
fig = plt.figure()

fig, (ax1, ax2) = plt.subplots(2, 1, gridspec_kw={'height_ratios': [2, 1]})


# Add an axis (subplot) to the figure
#ax1 = fig.add_subplot(1, 1, 1)
#ax2 = fig.add_subplot(2, 1, 1)

# why can't we do this, because ax1 already exists



# Plot some data on the axis
ax1.plot(x, y)
ax2.plot(x, z)

# Customize the plot
ax1.set_title('Top Plot')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')

ax2.set_title('Bottom Plot')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')

plt.subplots_adjust(hspace=0.5)

# Show the plot
plt.show()
```

```python
# here we use axes


# we create figure, we save figure in fig object
fig = plt.figure()

# use axes.  is the plural of axis

ax1 = fig.add_subplot(1, 2, 1)
ax1.set_title('x-y plot')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.plot(x, y)

# Subplot 2
ax2 = fig.add_subplot(1, 2, 2)
ax2.set_title('y-z plot')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')
ax2.plot(x, z)



# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

# Plot the data
plt.plot(x, y)

# Customize the tick marks with labels
plt.xticks([1, 2, 3, 4, 5], ['A', 'B', 'C', 'D', 'E'])  # Set tick positions and labels on the x-axis
plt.yticks([10, 20, 30, 40], ['Short', 'Medium', 'Tall', 'Very Tall'])  # Set tick positions and labels on the y-axis

# Add labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```
