```python
# here we use axes


# we create figure, we save figure in fig object
fig = plt.figure(figsize=(15,3))

# use axes.  is the plural of axis

ax1 = fig.add_subplot(1, 2, 1)
ax1.set_title('x-y plot')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.plot(x, y)

# Subplot 2
ax2 = fig.add_subplot(1, 2, 2)
ax2.set_title('y-z plot')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')
ax2.plot(x, z)



# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

# Plot the data
plt.plot(x, y)

# Customize the tick marks with labels
plt.xticks([1, 2, 3, 4, 5], ['A', 'B', 'C', 'D', 'E'])  # Set tick positions and labels on the x-axis
plt.yticks([10, 20, 30, 40], ['Short', 'Medium', 'Tall', 'Very Tall'])  # Set tick positions and labels on the y-axis

# Add labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

```python
a = np.random.normal(loc=0, scale=3, size=25)

b= np.round(a).astype(int)
c= b.reshape(5,5)

plt.imshow(c, cmap='viridis')

plt.colorbar()

for i in range(c.shape[0]):
    for j in range(c.shape[1]):
        plt.annotate(str(c[i, j]), xy=(j, i), ha='center', va='center', color='white')


plt.show()


```

```python
plt.hist(b)

counts, bins, _ = plt.hist(b)

for i in range(len(counts)):
    plt.text(bins[i] + 0.15, counts[i], str(int(counts[i])), ha='center', va='bottom')

plt.show()
```
