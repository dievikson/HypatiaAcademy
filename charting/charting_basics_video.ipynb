```python
# line

# scatter

import numpy as np


x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
import matplotlib.pyplot as plt

plt.plot(x,y)
plt.plot(x,z)

plt.show()

```

```python
#fig, (ax1, ax2) = plt.subplots(2, 1, gridspec_kw={'height_ratios': [2, 1]})

fig, (ax1, ax2) = plt.subplots(2, 1,)

ax1.plot(x,y)
ax2.plot(x,z)

plt.show()
```
