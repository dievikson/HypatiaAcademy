````python
import numpy as np

# cast in computers means to force a number to be a specific type

#

x = np.linspace(1,300,10).astype(np.int16)
y = 2 * x

z = 3 * x
print(x)

print("\n")

[x,y,z]
````

````python
import matplotlib.pyplot as plt

# Create a new figure
# this is the plain constructor to make a figure object.  meaning we don't it anything
fig = plt.figure()

fig, (ax1, ax2) = plt.subplots(2, 1, gridspec_kw={'height_ratios': [2, 1]})


# Add an axis (subplot) to the figure
#ax1 = fig.add_subplot(1, 1, 1)
#ax2 = fig.add_subplot(2, 1, 1)

# why can't we do this, because ax1 already exists



# Plot some data on the axis
ax1.plot(x, y)
ax2.plot(y, z)

# Customize the plot
ax1.set_title('Top Plot')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')

ax2.set_title('Bottom Plot')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')

# Show the plot
plt.show()
````
