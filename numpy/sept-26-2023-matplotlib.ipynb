```python
import numpy as np
import pandas as pd

study=np.array([9, 2, 6, 4,8])

results=np.array([100, 50, 80, 75, 85])


# weights
coefficient=6

# y-intercept
yintercept=43

prediction=coefficient*study+yintercept
prediction
```

```python

study=np.array([9, 2, 6, 4,8])

results=np.array([100, 50, 80, 75, 85])
prediction=np.array([97.3,55.1,79.2,67.1,91.3])

df=pd.DataFrame({
    "study": study,
    "results": results,
    "prediction": prediction
}
)





```

```python
# pandas when it has more than 1 column is called a dataframe
# when it has only one column it's called a series

df['study']
```

```python
coefficient,yintercept=np.polyfit(df['study'],df['results'],1)
print(coefficient)
print(yintercept)
```

```python
df['prediction']=np.array([97.3,55.1,79.2,67.1,91.3])
df

```

```python
# make a function to multiply prediction=study * coefficient + yintercept


def fsilly(d):
  return d['study'] * coefficient + yintercept



df['prediction']=df.apply(fsilly,axis=1)

df
```

```python
# now calculate the LSE for each row using the same approach

def fsilly(d):
  return (d['prediction']-d['results'])**2

df['LSE']=df.apply(fsilly,axis=1)
df
```

```python
total_lse = df['LSE'].sum(axis=0)

print(f"Total LSE: {total_lse}")

df['LSE'].mean(axis=0)
df['LSE'].std(axis=0)
```
