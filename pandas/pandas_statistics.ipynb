```python
import pandas as pd

# this data is downloaded from Kaggle

file="https://raw.githubusercontent.com/werowe/HypatiaAcademy/master/pandas/DataScience_salaries_2024.csv"
df = pd.read_csv(file)
df.head(100)
```

```python
# Here we show average (mean) salary in the US, France, and Germany
# here we tell it how to format the numbers into floating point numbers with 2 decimal positions
# otherwise it tends to use scientific notation where we don't want that
# we use reset_index to turn it back into a dataframe otherwise it would just be text and not nice format


pd.options.display.float_format = '{:.0f}'.format
pd.set_option('display.max_rows', None)


salaries=df[df['company_location'].isin(['US', 'FR', 'DE'])].groupby(['company_location','job_title'])['salary_in_usd'].mean().reset_index()

sorted=salaries.sort_values(["company_location", "salary_in_usd", "job_title"],ascending=False)

sorted
```

```python
# average of all salaries
df.loc[:,['salary']].mean()
```

```python
# average by country

salaries=df[(df['company_location'].isin(['US', 'FR', 'DE'])) & (df['work_year'] == 2023)].groupby(['company_location'])['salary_in_usd'].mean().reset_index()

sorted=salaries.sort_values(["company_location", "salary_in_usd"],ascending=False)

sorted
```

# Axis Defined

```python
# sum across row

df = pd.DataFrame({

      "a" : [1,2,3,4],
      "b" : [5,6,7,8]
})


```

The "axis" parameter in Pandas refers to the direction along which an operation is performed on data structures like Series and DataFrames. In a Series, which is a one-dimensional array of values, there is only "axis 0" representing the single dimension. On the other hand, a DataFrame, a two-dimensional data structure similar to a table, has both "axis 0" for rows and "axis 1" for columns.

```python


# add across rows

df.sum(axis=0)
```
